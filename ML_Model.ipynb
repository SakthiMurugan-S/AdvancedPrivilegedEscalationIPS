```python
import numpy as np
import joblib
import optuna
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression # Added as potential meta-learner
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, make_scorer
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification # Use a meaningful dataset generator
from sklearn.preprocessing import StandardScaler # Good practice for SVM
import warnings # To suppress potential warnings during optimization

# Suppress specific warnings if needed (e.g., XGBoost verbosity)
# warnings.filterwarnings("ignore", category=UserWarning)
# warnings.filterwarnings("ignore", category=FutureWarning)


# --- 1. Generate a More Realistic Simulated Dataset ---
print("Generating simulated dataset...")
n_samples = 5000
n_features = 20 # Total features
n_informative = 12 # Features that actually correlate with the target
n_redundant = 3 # Features that are linear combinations of informative ones
n_useless = n_features - n_informative - n_redundant # Noise features
# Create class imbalance (e.g., 70/30 split)
X, y = make_classification(
    n_samples=n_samples,
    n_features=n_features,
    n_informative=n_informative,
    n_redundant=n_redundant,
    n_clusters_per_class=2, # Makes it a bit harder
    weights=[0.7, 0.3], # Class imbalance
    flip_y=0.05, # Add some noise to labels
    random_state=42
)
print(f"Dataset shape: X={X.shape}, y={y.shape}")
print(f"Original class distribution: {np.bincount(y)}")

# --- 2. Split Data FIRST ---
# Stratify ensures the same proportion of classes in train and test sets
X_train_orig, X_test_orig, y_train_orig, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
print(f"Train shape: X={X_train_orig.shape}, y={y_train_orig.shape}")
print(f"Test shape: X={X_test_orig.shape}, y={y_test.shape}")
print(f"Test class distribution: {np.bincount(y_test)}")


# --- 3. Feature Scaling (Recommended, especially for SVM) ---
# Fit scaler ONLY on training data, then transform both train and test
print("Scaling features...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_orig)
X_test_scaled = scaler.transform(X_test_orig)
# Save the scaler along with the model if you deploy it
# joblib.dump(scaler, "scaler.pkl")


# --- 4. Feature Selection (Applied AFTER split, fitted ONLY on training data) ---
print("Performing feature selection...")
# Use a potentially stronger model for selection, tune its threshold if needed
selector = SelectFromModel(
    RandomForestClassifier(n_estimators=150, random_state=42, n_jobs=-1),
    threshold='median' # Keep features with importance > median importance
    # threshold=0.01 # Alternatively, use a small absolute threshold
)
# Fit on the scaled training data
selector.fit(X_train_scaled, y_train_orig)

X_train_selected = selector.transform(X_train_scaled)
X_test_selected = selector.transform(X_test_scaled) # Use same selector for test set

print(f"Selected {X_train_selected.shape[1]} features out of {X_train_scaled.shape[1]}")
# Save the selector if needed for deployment
# joblib.dump(selector, "selector.pkl")

# --- 5. Handle Class Imbalance using SMOTE (Applied AFTER split, ONLY on training data) ---
print("Applying SMOTE to training data...")
smote = SMOTE(random_state=42)
# Apply SMOTE to the *selected* and *scaled* training features
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_selected, y_train_orig)

print(f"Original training distribution: {np.bincount(y_train_orig)}")
print(f"Resampled training distribution: {np.bincount(y_train_resampled)}")
print(f"Resampled train shape: X={X_train_resampled.shape}, y={y_train_resampled.shape}")


# --- 6. Hyperparameter Tuning using Optuna with Cross-Validation ---
# Define accuracy scorer for Optuna
accuracy_scorer = make_scorer(accuracy_score)
N_TRIALS_OPTUNA = 30 # Increase trials for better search (adjust as needed)
CV_FOLDS = 5 # Number of folds for cross-validation within Optuna

# --- Objective Function for Optuna (using CV) ---
def objective(trial, model_name, X_train, y_train):
    # Common parameters for XGBoost
    xgb_common_params = {
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'use_label_encoder': False,
        'random_state': 42
    }

    if model_name == "xgb":
        params = {
            **xgb_common_params, # Include common params
            'n_estimators': trial.suggest_int("xgb_n_estimators", 100, 1000, log=True),
            'max_depth': trial.suggest_int("xgb_max_depth", 3, 15),
            'learning_rate': trial.suggest_float("xgb_learning_rate", 0.005, 0.3, log=True),
            'subsample': trial.suggest_float("xgb_subsample", 0.5, 1.0),
            'colsample_bytree': trial.suggest_float("xgb_colsample_bytree", 0.5, 1.0),
            'gamma': trial.suggest_float("xgb_gamma", 0, 5),
            'lambda': trial.suggest_float("xgb_lambda", 1e-8, 1.0, log=True), # L2 reg
            'alpha': trial.suggest_float("xgb_alpha", 1e-8, 1.0, log=True), # L1 reg
        }
        model = XGBClassifier(**params)

    elif model_name == "rf":
        params = {
            'n_estimators': trial.suggest_int("rf_n_estimators", 100, 1000, log=True),
            'max_depth': trial.suggest_int("rf_max_depth", 5, 50),
            'min_samples_split': trial.suggest_int("rf_min_samples_split", 2, 20),
            'min_samples_leaf': trial.suggest_int("rf_min_samples_leaf", 1, 20),
            'max_features': trial.suggest_categorical("rf_max_features", ['sqrt', 'log2', 0.5, 0.7]), # Test different strategies
            'criterion': trial.suggest_categorical("rf_criterion", ["gini", "entropy"]),
            'random_state': 42,
            'n_jobs': -1
        }
        model = RandomForestClassifier(**params)

    elif model_name == "svm":
        # *** CORRECTED SVM BLOCK ***
        # Suggest kernel first
        kernel_choice = trial.suggest_categorical("svm_kernel", ['rbf', 'poly', 'sigmoid'])

        # Define base parameters (without degree initially)
        params = {
            'C': trial.suggest_float("svm_C", 1e-3, 1e3, log=True),
            'kernel': kernel_choice, # Use the chosen kernel
            'gamma': trial.suggest_categorical("svm_gamma", ['scale', 'auto']), # Common gamma settings
            'probability': True, # Needed for ROC AUC and stacking predict_proba
            'random_state': 42
        }

        # Conditionally add degree if kernel is 'poly'
        if kernel_choice == 'poly':
            params['degree'] = trial.suggest_int("svm_degree", 2, 5)

        # You could add specific gamma tuning here if needed, e.g.:
        # if params['gamma'] == 'specific_value': # If you add another category like 'specific_value'
        #    params['gamma'] = trial.suggest_float("svm_gamma_val", 1e-4, 1e-1, log=True)

        model = SVC(**params)
        # *** END OF CORRECTED SVM BLOCK ***

    elif model_name == "final_xgb": # Tuning the meta-learner
         params = {
            **xgb_common_params, # Include common params
            'n_estimators': trial.suggest_int("final_xgb_n_estimators", 50, 500, log=True),
            'max_depth': trial.suggest_int("final_xgb_max_depth", 2, 8),
            'learning_rate': trial.suggest_float("final_xgb_learning_rate", 0.01, 0.2, log=True),
            'subsample': trial.suggest_float("final_xgb_subsample", 0.6, 1.0),
            'colsample_bytree': trial.suggest_float("final_xgb_colsample_bytree", 0.6, 1.0),
        }
         model = XGBClassifier(**params)

    else:
        raise ValueError(f"Unknown model_name: {model_name}")

    # --- Use cross-validation on the training set for robust evaluation ---
    cv = StratifiedKFold(n_splits=CV_FOLDS, shuffle=True, random_state=42)
    # Use the resampled training data for tuning
    # Using n_jobs=-1 within cross_val_score can speed things up if the model supports parallel CV
    # Be mindful of total CPU usage if Optuna is also parallelized (n_jobs > 1 in study.optimize)
    scores = cross_val_score(model, X_train, y_train, cv=cv, scoring=accuracy_scorer, n_jobs=1) # Use n_jobs=1 here if Optuna runs in parallel
    return np.mean(scores) # Return the average accuracy across folds

# --- Run Optuna Studies ---
# Consider using a persistent storage if runs are long or might be interrupted
# storage_name = "sqlite:///optimization.db"
# study_xgb = optuna.create_study(direction="maximize", study_name="XGBoost Optimization", storage=storage_name, load_if_exists=True)

print(f"\nOptimizing Base Models (Trials={N_TRIALS_OPTUNA}, CV Folds={CV_FOLDS})...")

# Run studies sequentially to avoid potential resource contention if n_jobs=-1 used heavily
study_xgb = optuna.create_study(direction="maximize", study_name="XGBoost Optimization")
study_xgb.optimize(lambda trial: objective(trial, "xgb", X_train_resampled, y_train_resampled), n_trials=N_TRIALS_OPTUNA, show_progress_bar=True)
best_params_xgb = study_xgb.best_params
print(f"Best XGB params: {best_params_xgb}, Best CV Accuracy: {study_xgb.best_value:.4f}")

study_rf = optuna.create_study(direction="maximize", study_name="RandomForest Optimization")
study_rf.optimize(lambda trial: objective(trial, "rf", X_train_resampled, y_train_resampled), n_trials=N_TRIALS_OPTUNA, show_progress_bar=True)
best_params_rf = study_rf.best_params
print(f"Best RF params: {best_params_rf}, Best CV Accuracy: {study_rf.best_value:.4f}")

study_svm = optuna.create_study(direction="maximize", study_name="SVM Optimization")
study_svm.optimize(lambda trial: objective(trial, "svm", X_train_resampled, y_train_resampled), n_trials=N_TRIALS_OPTUNA, show_progress_bar=True)
best_params_svm = study_svm.best_params
print(f"Best SVM params: {best_params_svm}, Best CV Accuracy: {study_svm.best_value:.4f}")

# --- Tune the Meta-Learner (Optional but recommended) ---
N_TRIALS_META = N_TRIALS_OPTUNA // 2 # Fewer trials might suffice
print(f"\nOptimizing Meta-Learner (Trials={N_TRIALS_META}, CV Folds={CV_FOLDS})...")
study_final_xgb = optuna.create_study(direction="maximize", study_name="Final XGB Optimization")
# Using the resampled data again - assumes meta-learner benefits from balanced data
study_final_xgb.optimize(lambda trial: objective(trial, "final_xgb", X_train_resampled, y_train_resampled), n_trials=N_TRIALS_META, show_progress_bar=True)
best_params_final_xgb = study_final_xgb.best_params
print(f"Best Final XGB params: {best_params_final_xgb}, Best CV Accuracy: {study_final_xgb.best_value:.4f}")


# --- 7. Define Base Models with Tuned Hyperparameters ---
# Common XGBoost params for instantiation
xgb_final_common_params = {
    'objective': 'binary:logistic',
    'eval_metric': 'logloss',
    'use_label_encoder': False,
    'random_state': 42
}

# Re-map parameter names from Optuna suggestions to model arguments
xgb_model = XGBClassifier(
    **xgb_final_common_params,
    n_estimators=best_params_xgb['xgb_n_estimators'],
    max_depth=best_params_xgb['xgb_max_depth'],
    learning_rate=best_params_xgb['xgb_learning_rate'],
    subsample=best_params_xgb['xgb_subsample'],
    colsample_bytree=best_params_xgb['xgb_colsample_bytree'],
    gamma=best_params_xgb['xgb_gamma'],
    reg_lambda=best_params_xgb.get('xgb_lambda', 1.0), # Use get with default if not suggested
    reg_alpha=best_params_xgb.get('xgb_alpha', 0.0)    # Use get with default if not suggested
)

rf_model = RandomForestClassifier(
    random_state=42, n_jobs=-1,
    n_estimators=best_params_rf['rf_n_estimators'],
    max_depth=best_params_rf['rf_max_depth'],
    min_samples_split=best_params_rf['rf_min_samples_split'],
    min_samples_leaf=best_params_rf['rf_min_samples_leaf'],
    max_features=best_params_rf['rf_max_features'],
    criterion=best_params_rf['rf_criterion']
)

# Handle potential SVM param variations
svm_params_final = {
    'probability': True, 'random_state': 42,
    'C': best_params_svm['svm_C'],
    'kernel': best_params_svm['svm_kernel'],
    'gamma': best_params_svm['svm_gamma']
}
# Add degree only if kernel is poly and degree was suggested
if svm_params_final['kernel'] == 'poly' and 'svm_degree' in best_params_svm:
    svm_params_final['degree'] = best_params_svm['svm_degree']

svm_model = SVC(**svm_params_final)


# Define Tuned Meta-Learner
final_xgb_model = XGBClassifier(
    **xgb_final_common_params,
    n_estimators=best_params_final_xgb['final_xgb_n_estimators'],
    max_depth=best_params_final_xgb['final_xgb_max_depth'],
    learning_rate=best_params_final_xgb['final_xgb_learning_rate'],
    subsample=best_params_final_xgb['final_xgb_subsample'],
    colsample_bytree=best_params_final_xgb['final_xgb_colsample_bytree']
)
# Alternative simpler meta-learner (often works well):
# final_lr_model = LogisticRegression(max_iter=1000, random_state=42, solver='liblinear')


# --- 8. Stacking Classifier ---
# Use cv=StratifiedKFold to generate out-of-fold predictions for the meta-learner training
# This prevents the base models from seeing the same data they predict on when training the meta-learner.
stacking_cv = StratifiedKFold(n_splits=CV_FOLDS, shuffle=True, random_state=42)

stacking_model = StackingClassifier(
    estimators=[
        ('rf', rf_model),
        ('xgb', xgb_model),
        ('svm', svm_model)
    ],
    # final_estimator=final_lr_model, # Try Logistic Regression first
    final_estimator=final_xgb_model, # Use the tuned XGBoost
    cv=stacking_cv, # Use CV for meta-learner training
    n_jobs=-1, # Use multiple cores for fitting base estimators if they support it
    passthrough=False # Set True if you want original features also passed to meta-learner (might require more tuning)
)

# --- 9. Train the Stacking Model ---
print("\nTraining final Stacking model...")
# Train on the resampled (selected and scaled) training data
stacking_model.fit(X_train_resampled, y_train_resampled)
print("Training complete.")

# --- 10. Predictions and Evaluation on the UNSEEN Test Set ---
print("\nEvaluating model on the test set...")
# Use the *selected* and *scaled* test features (NOT resampled)
y_pred = stacking_model.predict(X_test_selected)
y_prob = stacking_model.predict_proba(X_test_selected)[:, 1] # Probabilities for ROC-AUC

# Performance Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
try:
    # Ensure there are both classes present in y_test and probabilities exist
    if len(np.unique(y_test)) > 1 and hasattr(stacking_model, "predict_proba"):
         roc_auc = roc_auc_score(y_test, y_prob)
    else:
         roc_auc = 0.0
         print("Warning: ROC AUC score could not be calculated (single class in y_test or probabilities unavailable).")
except ValueError as e:
    roc_auc = 0.0 # Handle other potential errors
    print(f"Warning: ROC AUC score could not be calculated. Error: {e}")


# Display results
print(f"\nOptimized and Tuned Stacked Model Performance Metrics (Test Set):")
print(f"Accuracy:  {accuracy * 100:.2f}%")
print(f"Precision: {precision * 100:.2f}%")
print(f"Recall:    {recall * 100:.2f}%")
print(f"F1-Score:  {f1 * 100:.2f}%")
print(f"ROC-AUC:   {roc_auc * 100:.2f}%")

# --- 11. Confusion Matrix ---
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Predicted Normal (0)", "Predicted Anomaly (1)"], # Adjust labels as needed
            yticklabels=["True Normal (0)", "True Anomaly (1)"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Tuned Stacking Model (Test Set)")
plt.tight_layout() # Adjust layout
plt.show()

# --- 12. Save Final Model ---
# Save the entire pipeline (including scaler, selector if used outside pipeline object)
# For simplicity here, just saving the trained stacking model.
# In practice, consider using scikit-learn's Pipeline object to chain steps.
model_filename = "tuned_stacking_model.pkl"
joblib.dump(stacking_model, model_filename)
print(f"\nOptimized Stacking Model saved as {model_filename}")

# Remember to also save the scaler and selector if needed for future predictions on raw data
scaler_filename = "scaler.pkl"
joblib.dump(scaler, scaler_filename)
print(f"Scaler saved as {scaler_filename}")

selector_filename = "selector.pkl"
joblib.dump(selector, selector_filename)
print(f"Feature selector saved as {selector_filename}")

print("\nScript finished successfully.")
```
